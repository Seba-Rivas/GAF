## P1

```python
import numpy as np
import matplotlib.pyplot as plt
import math

L=0.8
l=0.4
t=np.arange(1,50,1)
h=lambda x: L*l*( (math.exp(-L*x)- math.exp(-l*x)) / (l*math.exp(-L*x) - L*math.exp(-l*x)) )
H=[]
for i in range(len(t)):
    H.append(h(t[i]))
plt.plot(t,H)
plt.xlabel('Tiempo')
plt.ylabel('h(t)')
plt.show()
```

## P2

```python
h2 = lambda y: y/(1+y)
H2=[]
for i in range(len(t)):
    H2.append(h2(t[i]))
plt.plot(t,H2)
plt.xlabel('Tiempo')
plt.ylabel('h(t)')
plt.show()
```

## P4

```python
k=5
p=0.1
h3 = lambda x: ( p*L*math.exp(-L*x) + (1-p)*( 1/(L**5*math.gamma(k)) ) * x**(k-1) * math.exp(-L*x) )/(p*math.exp(-L*x) + (1-p)*math.exp(-L*x)*(1+x*L+((x*L)**2)/2+((x*L)**3)/6+ ((x*L)**4)/24 ) )
H3=[]
for i in range(len(t)):
    H3.append(h3(t[i]))
plt.plot(t,H3)
plt.xlabel('Tiempo')
plt.ylabel('h(t)')
plt.show()
```

## P6

```python
Lo = 6.5*10**(-4)
L_ = 3.5*10**(-4)
p = np.arange(0, 1+0.001, 0.001)
lbda = lambda p: Lo + 2*p*(p-1.6)*L_
Lbda = list()
index =list()
for i in range(len(p)):
    Lbda.append(lbda(p[i]))
    if p[i] == 0.4 or p[i] == 0.8 or p[i] == 1:
        index.append(i)        

plt.plot(p,Lbda)
plt.scatter(p[400], Lbda[400], label='$\lambda_p(0.4)$={}'.format(round(Lbda[index[0]],5)), color='green')
plt.scatter(p[800], Lbda[800], label='$\lambda_p(0.8)$={}'.format(round(Lbda[index[1]],8)), color='red')
plt.scatter(p[1000], Lbda[1000], label='$\lambda_p(1)$={}'.format(round(Lbda[index[2]],5)), color='yellow')
plt.xlabel('Tiempo')
plt.ylabel('$\lambda_p$')
plt.legend()
plt.show()
```
